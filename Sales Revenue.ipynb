## Sales Revenue

This example will use pandas to process sales data.

```python
import pandas as pd
```

```python
# load sales calls data
sales_calls = pd.read_csv('data/sales_calls.csv')
print(sales_calls)
print(type(sales_calls))
```

```python
# load sales revenue data
sales_revenue = pd.read_csv('data/sales_revenue.csv')
print(sales_revenue)
```

```python
# merge both dataframes
calls_revenue = pd.merge(sales_calls, sales_revenue, on=['Territory', 'Month'])
print(calls_revenue)
```

```python
# filtering - select rows for territory 1 only
print(calls_revenue[calls_revenue.Territory == 1])
```

```python
# select rows for month 4
print(calls_revenue[calls_revenue['Month'] == 4])
```

```python
# TRY: select records for team member Ali
print(calls_revenue[calls_revenue['Team member'] == 'Ali'])
```

```python
# TRY: select records where amount per call is greater than 500
print(calls_revenue[calls_revenue.Amount/calls_revenue.Calls > 500])
```

```python
print(calls_revenue[calls_revenue['Amount']/calls_revenue['Calls'] > 500])
```

```python
# calculate Amount per Calls and store in new column in dataframe
calls_revenue['Amount_Calls'] = calls_revenue.Amount/calls_revenue.Calls
print(calls_revenue)
```

```python
print(calls_revenue[calls_revenue.Amount_Calls > 500])
```

```python
print("Max Calls: {}".format(calls_revenue.Calls.max()))
print("Min Calls: {}".format(calls_revenue.Calls.min()))
```

```python
# display statistics for dataframe
calls_revenue.describe()
```

```python
# display selected columns statistics
calls_revenue[['Calls', 'Amount', 'Amount_Calls']].describe()
```

```python
# get total calls
print(calls_revenue.Calls.sum())
```

```python
# groupby territory - get the total calls and amount
print(calls_revenue[['Territory', 'Calls', 'Amount']]
      .groupby(['Territory']).sum())
```
