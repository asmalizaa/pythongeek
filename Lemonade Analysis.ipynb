## Lemonade Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data
data = pd.read_excel('data/Lemonade.xlsx')
data
```

```python
# get column names
cols = data.columns
print(cols)
```

```python
# get the first 5 rows
data.head()
```

```python
# get the first 10 rows
data.head(10)
```

```python
# get the last 5 rows
data.tail()
```

```python
# drop columns
data = data.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'])
```

```python
data
```

```python
data.describe()
```

```python
# find the date/day when the sales is highest
data[data.Sales == data.Sales.max()]
```

```python
# find the date/day when the sales is lowest
data[data.Sales == data.Sales.min()]
```

```python
# find correlation between Temperature and Sales
data['Temperature'].corr(data['Sales'])
```

```python
# plot data to see correlation between Temperature and Sales
plt.title('Sales vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.grid()
plt.scatter(data['Temperature'], data['Sales'])
plt.show()
```

```python
# find correlation between Rainfall and Sales
data['Rainfall'].corr(data['Sales'])
```

```python
# plot data to see correlation between Rainfall and Sales
plt.title('Sales vs Rainfall')
plt.xlabel('Rainfall')
plt.ylabel('Sales')
plt.grid()
plt.scatter(data['Rainfall'], data['Sales'])
plt.text(1, 30, 'y = mx + c', fontsize=12, fontweight='bold', color='red')
plt.show()
```
