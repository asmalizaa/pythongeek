## Read Users

This example will read text file.

```python
# set the file path
file = '../data/users.txt'

# open the file for reading
data = open(file)

# reading the contents
for line in data :
    #print(line.strip())
    
    # extract values for each line
    values = line.strip().split(':')
    #print(values)
    
    # format the output
    #print("{} {} {} {} {}".format(values[0], values[1], values[2], values[3], values[4]))
    print(values[0].ljust(10), values[1].ljust(20), values[2].ljust(20), values[3].ljust(20), values[4])
    
```
